```python
from typing import List
```

```python
def containsDuplicate(nums: List[int]) -> bool:
    if len(set(nums)) < 1:
        return False
    from collections import Counter
    nums_count = Counter(nums).most_common(1)
    if nums_count[0][1] > 1:
        return True
    else:
        return False
```

```python
containsDuplicate([1,2,3,1])
```
